## 1 Drifit Diffusion Model (5 points)

Use whatever analytical methods to check if Tren's super EZ implementation(s) of the drift diffusion model produce exGaussian data, especially the exponential long tail. Remember to play around with different model parameters to see how they affect simulation results.

```python
# Tren's implementation
random.seed(1)

mu,sigma=0.2,2 # Diffusion type1
p=0.4 # Diffusion type2
Yend=20 # Threshold for response
Ntrial=1000

RTs=zeros(Ntrial)
for i in range(Ntrial):
    Y=0 # initial drift position
    Ys=[] # drift history
    RT=0 # Reset watch
    while(Y<Yend):
        dY=mu+random.normal(mu,sigma); # type1
        #dY=float((random.rand(1)-p)>0); # type2
        Y=Y+dY
        Ys.append(Y); # Diffusion history in each trial
        RT=RT+1 # clock is running
    RTs[i]=RT;
subplot(1,2,1);plot(Ys);# plot only the last trial to see the diffusion process
subplot(1,2,2);hist(RTs);
```

#### (0) Estimate parameters & Draw plot

```python
from scipy.stats import norm,exponnorm
from numpy import linspace
from pylab import plot,show,hist,figure,title

param = exponnorm.fit(RTs) # distribution fitting
 
# the standard deviation of the fitted distribution
x = linspace(-5,300,1000)
# fitted distribution
pdf_fitted = exponnorm.pdf(x,*param[0:-2],loc=param[1],scale=param[2])

title('exGaussian distribution')
#plot(x,pdf_fitted,'r-',x,pdf,'b-')
plot(x,pdf_fitted,'r-')
hist(RTs,normed=1,alpha=.3,color='g')
show()
```

#### (1) Do Monte Carlo to compare the exGaussian Distribution

```python
from scipy import stats
nMC=1000;
alpha=0.05;
np.random.seed(seed=2)

def getQuantile(data,K,loc,scale):
    n= len(data)
    gProb = (linspace(1,n,n)-0.5)/n # % assign prob.
    pQuantile = exponnorm.ppf(gProb,K,loc,scale)
    zData = stats.zscore(data)
    zQuantile=sorted(zData)
    return pQuantile,zQuantile

MC_cor=[]
alpha=0.05
for i in range(nMC):
    exGtemp = exponnorm.rvs(param[0], param[1], param[2], size=Ntrial)
    exG_pQ, exG_zQ = getQuantile(exGtemp,param[0],param[1],param[2])
    MC_cortemp = corrcoef(exG_pQ, exG_zQ)
    #print(qCorrT[0,1])
    MC_cor.append(MC_cortemp[0,1])        
    MC_cor=sorted(MC_cor)
    
criticalValue=MC_cor[int(nMC*alpha-1)]
print('CV =',criticalValue)
Data_pQ, Data_zQ = getQuantile(RTs,param[0],param[1],param[2])
DataCor = corrcoef(Data_pQ,Data_zQ)[0,1]
print('DataCor =',DataCor)
#print(qCorr)
if (DataCor < criticalValue):
    print("Data generated from tren\'s implementation(s) is probably not a exGaussian Distribution.")
else:
    print("Data generated from tren\'s implementation(s) is probably a exGaussian Distribution.")
```

```python
def EZdata(v,a,Ter,s):
    #s = np.zeros(shape=(len(v)))
    s2 = s**2 ;
    
    #-- Probability correct
    y = (-v * a )/ s2 ;
    Pc = 1 / (1+exp(y)) ;
    
    #-- Postpone special case v=0
    idx = (v!=0)

    vi = v[idx] ;
    yi = y[idx] ;
    ai = a
    
    #if (len(a)==1):
    #    ai = a
    #else :
    #    ai = np.zeros(shape=(len(v)))
    #    ai = a[idx]
    ai = a[idx]
    # allow for scalar a
    
    #-- Variance of response times -- Equation 6 in Wagenmakers et al
    exp_yi = np.exp(yi) ;
    VRTi = (2*yi*exp_yi - np.exp(2*yi) + 1) / (exp_yi+1)**2 ;
    VRTi = VRTi * (ai*s2/(2*vi**3)) ;
    VRT = np.zeros(shape=(len(v)))
    VRT[idx] = VRTi
    
    #-- Mean decision time -- Equation 9
    MDTi = (1-exp_yi)/(1+exp_yi) ;
    MDTi = MDTi * (ai/(2*vi)) ;

                 
    MDT = np.zeros(shape=(len(v)))
    MDT[idx] = MDTi
    
    #-- Now handle special case v=0
    #idx = find(v==0) ;
    idx = (v==0)
    if (logical_not(np.all(idx==0))):
        if (len(a)==1):
            ai = a
        else:
            ai = a[idx]
        
        #- Variance of response times
        VRTi = ai**4 / (24*s2**2);     # Eq. on top of p. 9, right
        VRT = np.zeros(shape=(len(v)))
        VRT[idx] = VRTi
        #- Mean decision time
        MDTi = ai**2 / (4*s2) ;
        MDT = np.zeros(shape=(len(v)))
        MDT[idx] = MDTi ;

    
    #-- Observable Mean RT = Mean decision time + Ter
    MRT = MDT + Ter ;   # Equation 8
    
    #%---   Return Pc, VRT, MRT
    #%%%%%% End of file EZdata.m
    return [Pc,VRT,MRT]

v = np.array([.1])
a = np.array([0.14])
Ter = np.array([0.3])
s = np.array([0.1])

result = EZdata(v, a, Ter, s)
print("Pc:",result[0])
print("VRT:",result[1])
print("MRT:",result[2])
```
